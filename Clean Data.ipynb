```python
import pandas as pd
```

```python
hot_stuff = pd.read_csv('Resources/Hot_Stuff.csv')
hot_stuff.tail()
```

```python
weeks = hot_stuff.sort_values("Weeks on Chart", ascending=False)
```

```python
weeks
```

```python
result_df = weeks.drop_duplicates(subset=['Song'])
```

```python
result_df
```

```python
features = pd.read_csv('Resources/Audio_Features.csv')
```

```python
features
```

```python
merged = result_df.merge(features, left_on='SongID', right_on="SongID")
```

```python
merged
```

```python
merged.sort_values("Peak Position")
```

```python
noNANS = merged.dropna(how='any')
```

```python
noNANS
```

```python
noNANS.to_csv(r'Path where you want to store the exported CSV file\File Name.csv')
```
